# **Bag of Words (BoW):**
Overview:

The Bag of Words model is a simple and commonly used technique to represent text data.
It treats each document as an unordered set of words, disregarding grammar and word order.
The model creates a vocabulary of unique words in the entire corpus and represents each document as a vector of word frequencies.

*Process:*

1. Tokenization: Break each document into individual words or tokens.
2. Vocabulary Building: Create a vocabulary of unique words from the entire corpus.
3. Vectorization: Represent each document as a vector, where each element corresponds to the frequency of a word in the vocabulary.

Example:
Consider three documents:

Document 1: "I love machine learning"

Document 2: "Machine learning is fun"

Document 3: "I love programming and machine learning"

Vocabulary: ["I", "love", "machine", "learning", "is", "fun", "programming", "and"]

Representation:


Document 1: [1, 1, 1, 1, 0, 0, 0, 0]

Document 2: [0, 0, 1, 1, 1, 1, 0, 0]

Document 3: [1, 1, 1, 1, 0, 0, 1, 1]


BoW is simple and interpretable but may lose semantic information.

TF-IDF stands for Term Frequency Inverse Document Frequency of records

tf(t,d) = count of t in d / number of words in d

df(t) = occurrence of t in documents

idf(t) = log(N/df(t))

tf-idf(t, d) = tf(t, d) * idf(t)



Example:
Consider three documents:

Review 1: "This movie is very scary and long"

Review 2: "This movie is not slow and is scary"

Review 3: "This movie is spooky and good"

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    "Thor eating pizza, Loki is eating pizza, Ironman ate pizza already",
    "Apple is announcing new iphone tomorrow",
    "Tesla is announcing new model-3 tomorrow",
    "Google is announcing new pixel-6 tomorrow",
    "Microsoft is announcing new surface tomorrow",
    "Amazon is announcing new eco-dot tomorrow",
    "I am eating biryani and you are eating grapes"
]
```

```python
#let's create the vectorizer and fit the corpus and transform them accordingly
v = TfidfVectorizer()
v.fit(corpus)
transform_output = v.transform(corpus)
#print(transform_output)
```

```python
#let's print the vocabulary

print(v.vocabulary_)
```

```python
#let's print the idf of each word:

all_feature_names = v.get_feature_names_out()

for word in all_feature_names:

    #let's get the index in the vocabulary
    indx = v.vocabulary_.get(word)

    #get the score
    idf_score = v.idf_[indx]

    print(f"{word} : {idf_score}")
```

```python
#let's print the transformed output from tf-idf
print(transform_output.toarray())
```

```python
import pandas as pd

#read the data into a pandas dataframe
df = pd.read_csv("ecommerceDataset.csv")
print(df.shape)
```

```python
df = df.set_axis(["label", "text"],
                    axis=1)
#check the distribution of labels
print(df['label'].value_counts())

new_df = df.dropna() # atleast one na then drop
new_df

df_percent = new_df.sample(frac=0.02)

df_percent.shape
print(df_percent['label'].value_counts())
```

```python
import spacy

# load english language model and create nlp object from it
nlp = spacy.load("en_core_web_sm")

def preprocess(text):
    # remove stop words and lemmatize the text
    doc = nlp(text)
    filtered_tokens = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(token.lemma_)

    return " ".join(filtered_tokens)
```

```python
df_percent['preprocessed_txt'] = df_percent['text'].apply(preprocess)
```

```python
print(df_percent.info())
```

```python
#Add the new column which gives a unique number to each of these labels

df_percent['label_num'] = df_percent['label'].map({
    'Household' : 0,
    'Books': 1,
    'Electronics': 2,
    'Clothing & Accessories': 3
})

#checking the results
df_percent.head(5)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df_percent.preprocessed_txt,
    df_percent.label_num,
    test_size=0.2, # 20% samples will go to test dataset
    random_state=2022,
    stratify=df_percent.label_num
)
```

```python
print("Shape of X_train: ", X_train.shape)
print("Shape of X_test: ", X_test.shape)
```

```python
#use CountVectorizer for pre-processing the text.

#use KNN as the classifier

#print the classification report.
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

#1. create a pipeline object
clf = Pipeline([
     ('vectorizer_count',CountVectorizer()),
     ('KNN', KNeighborsClassifier())
])

#2. fit with X_train and y_train
clf.fit(X_train, y_train)


#3. get the predictions for X_test and store it in y_pred
y_pred = clf.predict(X_test)


#4. print the classfication report
print(classification_report(y_test, y_pred))
```

```python
#use TF-IDF for pre-processing the text.

#use KNN as the classifier

#print the classification report.
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

#1. create a pipeline object
clf = Pipeline([
     ('vectorizer_tfidf',TfidfVectorizer()),
     ('KNN', KNeighborsClassifier())
])

#2. fit with X_train and y_train
clf.fit(X_train, y_train)


#3. get the predictions for X_test and store it in y_pred
y_pred = clf.predict(X_test)


#4. print the classfication report
print(classification_report(y_test, y_pred))
```

#Wordnet
WordNet is a lexical database for the English language, developed under the direction of George A. Miller and maintained by the Cognitive Science Laboratory at Princeton University.

1. Synsets (Synonym Sets): Collections of synonymous words or phrases that express the same concept.

Each synset represents a single concept and includes a list of synonymous words or phrases that share the same meaning.
Example: The synset for "dog" includes the words "dog," "domestic dog," and "Canis familiaris."


2. Hypernyms and Hyponyms: Words that represent a broader category (hypernyms) and more specific instances (hyponyms).

Hypernyms: More general terms. For example, "animal" is a hypernym of "dog."
Hyponyms: More specific terms. For example, "bulldog" is a hyponym of "dog."
These relationships form a hierarchical tree structure where more specific concepts (hyponyms) branch out from more general concepts (hypernyms).

3. Meronyms and Holonyms: Words that denote a part-whole relationship.

Holonyms: Whole-part relationships where a term represents a whole and its parts. For example, "tree" is a holonym of "branch."
Meronyms: Part-whole relationships where a term represents a part of a whole. For example, "branch" is a meronym of "tree."

4. Antonyms: Words with opposite meanings.
5. Definitions and Examples: Short definitions and example sentences for each word

```python
#Computing word Similarity

from nltk.corpus import wordnet as wn

# Synsets for two words
word1 = wn.synset('dog.n.01')
word2 = wn.synset('cat.n.01')

# Path Similarity
path_similarity = word1.path_similarity(word2)
print(f"Path similarity between 'dog' and 'cat': {path_similarity}")

# Wu-Palmer Similarity
wup_similarity = word1.wup_similarity(word2)
print(f"Wu-Palmer similarity between 'dog' and 'cat': {wup_similarity}")

# Leacock-Chodorow Similarity
lch_similarity = word1.lch_similarity(word2)
print(f"Leacock-Chodorow similarity between 'dog' and 'cat': {lch_similarity}")
```

#**Word to Vector (Word2Vec):**

Overview:

Word2Vec is a popular word embedding technique used to represent words in a continuous vector space. It was introduced by Tomas Mikolov and colleagues at Google in 2013. The key idea behind Word2Vec is to capture the semantic relationships between words based on their context in a large corpus of text. There are two main models in Word2Vec: Continuous Bag of Words (CBOW) and Skip-gram.

https://arxiv.org/pdf/1301.3781


*Process:*

1. Training: Train the Word2Vec model on a large corpus to learn word embeddings.
2. Vectorization: Represent each word as a dense vector in a continuous space.

Comparison:

BoW is simple and interpretable but may lose semantic information.
Word2Vec captures semantic relationships but requires training on large datasets.
Both techniques have their use cases depending on the requirements of the NLP task. BoW is often used for text classification, while Word2Vec is valuable for tasks involving semantic understanding and word relationships.

```python
from gensim.models import Word2Vec

# Example sentences
sentences = [
    "the quick brown fox jumps over the lazy dog".split(),
    "never jump over the lazy dog quickly".split(),
    "bright sun shines over the hill".split(),
    "the dog is bright and quick".split()
]

# Train Word2Vec model
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, sg=0)  # sg=0 for CBOW, sg=1 for Skip-Gram

# Get word vector
word_vector = model.wv['dog']
print("Word vector for 'dog':", word_vector)

# Find similar words
similar_words = model.wv.most_similar('dog')
print("Words similar to 'dog':", similar_words)
```

#Sentiment Analysis

Sentiment Analysis is the task of determining the sentiment (positive, negative or neutral) conveyed by a piece of text.

Example: Twitter data, Reviews, Social media

```python
import nltk
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from gensim.models import Word2Vec
from nltk.corpus import movie_reviews

# Download the dataset
nltk.download('movie_reviews')

# Load the dataset
def load_movie_reviews():
    documents = [(list(movie_reviews.words(fileid)), category)
                 for category in movie_reviews.categories()
                 for fileid in movie_reviews.fileids(category)]
    np.random.shuffle(documents)
    return documents

documents = load_movie_reviews()

# Prepare data
texts, labels = zip(*documents)
texts = [' '.join(text) for text in texts]
labels = np.array([1 if label == 'pos' else 0 for label in labels])

# Split data
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.2, random_state=42)
```

```python
# CountVectorizer
count_vectorizer = CountVectorizer()
X_train_counts = count_vectorizer.fit_transform(X_train)
X_test_counts = count_vectorizer.transform(X_test)

# Train Logistic Regression
clf_count = LogisticRegression(max_iter=200)
clf_count.fit(X_train_counts, y_train)

# Predict and Evaluate
y_pred_count = clf_count.predict(X_test_counts)
print("CountVectorizer Accuracy:", accuracy_score(y_test, y_pred_count))
print("Classification Report:\n", classification_report(y_test, y_pred_count))
```

```python
# TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Train Logistic Regression
clf_tfidf = LogisticRegression(max_iter=200)
clf_tfidf.fit(X_train_tfidf, y_train)

# Predict and Evaluate
y_pred_tfidf = clf_tfidf.predict(X_test_tfidf)
print("TF-IDF Vectorizer Accuracy:", accuracy_score(y_test, y_pred_tfidf))
print("Classification Report:\n", classification_report(y_test, y_pred_tfidf))
```

```python
# Tokenize sentences for Word2Vec
X_train_tokens = [text.split() for text in X_train]
X_test_tokens = [text.split() for text in X_test]

# Train Word2Vec model
w2v_model = Word2Vec(sentences=X_train_tokens, vector_size=100, window=5, min_count=1, workers=4)
```

```python
# Average Word Vectors for each document
def document_vector(doc):
    # Remove out-of-vocabulary words
    doc = [word for word in doc if word in w2v_model.wv.index_to_key]
    return np.mean(w2v_model.wv[doc], axis=0)

X_train_w2v = np.array([document_vector(doc) for doc in X_train_tokens])
X_test_w2v = np.array([document_vector(doc) for doc in X_test_tokens])
```

```python
# Train Logistic Regression
clf_w2v = LogisticRegression(max_iter=10000)
clf_w2v.fit(X_train_w2v, y_train)

# Predict and Evaluate
y_pred_w2v = clf_w2v.predict(X_test_w2v)
print("Word2Vec Accuracy:", accuracy_score(y_test, y_pred_w2v))
print("Classification Report:\n", classification_report(y_test, y_pred_w2v))
```
